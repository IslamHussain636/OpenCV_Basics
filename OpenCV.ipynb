```python
# basic functions and manipulation in opencv

import cv2 as cv 
import numpy as np 


img=cv.imread('data_for_cv/islam.jpeg')
img=cv.resize(img,(450,450))
 

# resizing an Image
resized_img= cv.resize(img, (250,250))

# gray Image 
gray_img= cv.cvtColor(img, cv.COLOR_BGR2GRAY)

#B&W
(thresh, binary)= cv.threshold(img, 127,255, cv.THRESH_BINARY)

# Blur
blur_img= cv.GaussianBlur(img,(7,7),0) # (23x23)

#edge detection
edge_img= cv.Canny(img,53,48)

# thickness of line
mat_kernal= np.ones((3,3),np.uint8)
dilated_img= cv.dilate(edge_img,(mat_kernal),iterations=1)

# errosion thiner
err_img= cv.erode(dilated_img, mat_kernal, iterations=1)

# cropping we will use 
print('The size of our image is:',img.shape)
crop_img=img[0:200, 200:300]


cv.imshow('Orignal',img)
cv.imshow('resizes',resized_img)
cv.imshow('B&W',binary)
cv.imshow('gray',gray_img)
cv.imshow('blur',blur_img)
cv.imshow('edge',edge_img)
cv.imshow('dilated',dilated_img)
cv.imshow('errored',err_img)
cv.imshow('cropped',crop_img)
cv.waitKey(0)
cv.destroyAllWindows()

```

```python
#straghting a scan document
#changing the perspective 
import cv2 as cv 
import numpy as np 

img =cv.imread('islam.JPEG')

#defining points 
point1= np.float32([[233,196],[82,471],[522,169],[715,482]])

width=800
height=900

point2 = np.float32([[0,0],[width,0],[0,height],[width,height]])

matrix= cv.getPerspectiveTransform(point1,point2)

out_img= cv.warpPerspective(img,matrix,(width, height))
# cv.resize()
cv.imshow('orignal',img)
cv.imshow('transform',out_img)
cv.imwrite('data_for_cv/wrap.png')

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# how to detect specific color inside python in pics 

import cv2 as cv 
import numpy as np 

img=cv.imread('islam.JPEG')
img=cv.resize(img, 250,250)

#convert into HSV (Hue, Saturation, value)
#hsv_img= cv.cvtColor(img,cv.COLOR_BGR2HSV)

#slider

def slider():
    pass

path= 'islam.JPEG'


cv.namedWindow("Bars")
cv.resizeWindow("Bars",900,300)

cv.createTrackbar('Hue Min', 'Bars', 0,179,slider)
cv.createTrackbar('Hue Max', 'Bars', 179,179,slider)
cv.createTrackbar('Sat Min', 'Bars', 0,255,slider)
cv.createTrackbar('Sat Max', 'Bars', 255,255,slider)
cv.createTrackbar('Val Min', 'Bars', 0,255,slider)
cv.createTrackbar('Val Max', 'Bars', 255,255,slider)

while True:
      img= cv.imread(path)
      hsv_img= cv.cvtColor(img, cv.COLOR_BGR2HSV)

      hue_min=cv.getTrackbarPos('Hue Min', 'Bars')
      hue_max=cv.getTrackbarPos('Hue Max', 'Bars')
      sat_min=cv.getTrackbarPos('Sat Min', 'Bars')
      sat_max=cv.getTrackbarPos('Sat Max', 'Bars')
      val_min=cv.getTrackbarPos('Val Min', 'Bars')
      val_max=cv.getTrackbarPos('Val Max', 'Bars')
      
      print(hue_min,hue_max,sat_min,sat_max,val_min,val_max)
# to see changes inside images
      lower= np.array[hue_min,sat_min,sat_min]
      upper= np.array[hue_max,sat_max,val_max]

      mask_img=cv.inRange(hsv_img,lower,upper)
      out_img=cv.bitwise_and(img,img,mask=mask_img)
      cv.imshow('Original', img)
      cv.imshow('HSV', hsv_img)
      cv.imshow('mask', mask_img)
      cv.imshow('final', out_img)
      if cv.waitKey(1) & 0xff == ord('q'):
          break 

cv.destroyAllWindows()
```

```python
# face detection in Images 

import cv2 as cv

face_cascade=cv.CascadeClassifier('haarcascade_frontalface_default.xml')

img= cv.imread('islam.JPEG')
img=cv.resize(img,(300,300))
gray_img=cv.cvtColor(img,cv.COLOR_BGR2GRAY)

faces=face_cascade.detectMultiScale(gray_img,1.1,4)

for (x,y,w,h) in faces:
    cv.rectangle(img,(x,y), (x+w, y+h), (255,0,0),2)


cv.imshow("Gray image",gray_img)
cv.imshow("image",img)
cv.imwrite('data-for_cv/islam.png',img)
cv.waitKey(0)
cv.destroyAllWindows()
```
